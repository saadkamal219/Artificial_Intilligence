**Simpe Graph with given coordinates**

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

**Types of ploting**

```python
x = np.array([1,5,11,6,8,0])
y = np.array([6,9,7,3,33,50])

plt.plot(x,y,'o')
plt.show()

plt.plot(x,y,'*')
plt.show()

plt.plot(x,y,marker='_') #just the same
plt.show()

plt.plot(x,y, 'o:r') #line formating
plt.show()

plt.plot(x,y, 'o-.r')
plt.show()

plt.plot(x,y, 'o-r', ms = 20) # ms refers marker size
plt.show()

plt.plot(x,y, 'o-g', ms = 20, mec = 'r') # mec colors edge of the marker
plt.show()

plt.plot(x,y, 'o-r', ms = 20, mfc = 'w') # mfc colors the inner marker
plt.show()
```

**Types of lines**

```python
x = np.array([1,5,11,6,8,0])
y = np.array([6,9,7,3,33,50])

plt.plot(x,y, linestyle = 'dotted') # ls = ':' another is ls = '--' means 'dashed'
plt.show()
```

**Lables**

```python
x = np.array([1,5,11,6,8,0])
y = np.array([6,9,7,3,33,50])

font1 = {'family':'serif','color':'black','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x,y, marker = 'o', ms = 10, mec = 'k', mfc = 'y', ls = '--')

plt.title('Pizza to Coca-Cola Production Ratio', fontdict = font1, loc = 'right') #loc defines the location of the title
plt.xlabel('Pizza (M)', fontdict = font2)
plt.ylabel('Coca-Cola (M)', fontdict = font2)

plt.show()
```

**Grid**

```python
x = np.array([1,5,11,6,8,0])
y = np.array([6,9,7,3,33,50])

font1 = {'family':'serif','color':'black','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x,y, marker = 'o', ms = 10, mec = 'k', mfc = 'y', ls = '--')

plt.grid(axis = 'both') #to print just x-axis grids we just need change the value of the axis in grid function
                        #do same to print the y-axis grids
                        #default value is both
plt.title('Pizza to Coca-Cola Production Ratio', fontdict = font1, loc = 'right') #loc defines the location of the title
plt.xlabel('Pizza (M)', fontdict = font2)
plt.ylabel('Coca-Cola (M)', fontdict = font2)

plt.show()
```

**Scatter**

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'red')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#000000')

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c = colors, cmap = 'viridis')

plt.colorbar()

plt.show()
```

**Bar**

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([10, 95, 75, 50])


plt.bar(x, y, color = 'yellow', width = .3)
plt.show()
plt.barh(x, y, color = 'red', height = .5)
plt.show() #h means horizontal
```

**Histogram**

```python
random = np.random.normal(170, 10, 250)

plt.hist(random, color = 'red')
plt.show()
```

**Pie Chart**

```python
x = np.array([35, 25, 25, 15])

mylabels = ['Code', 'Sleep', 'DIU', 'Others']
myexplode = [0.1, 0, 0, 0]
colors = ['blue', 'black', 'green', 'red']

plt.title('Work Percentage')
plt.pie(x, labels = mylabels)
plt.show()

plt.title('Work Percentage')
plt.pie(x, labels = mylabels, startangle = 180)
plt.show()

plt.title('Work Percentage')
plt.pie(x, labels = mylabels, startangle = 90, explode = myexplode, shadow = True)
plt.show()

plt.title('Work Percentage')
plt.pie(x, labels = mylabels, startangle = 0, explode = myexplode, shadow = True, colors = colors)
plt.legend(title = 'Routine: ')
plt.show()
```
