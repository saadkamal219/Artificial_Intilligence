```python
import pandas as pd
from sklearn.metrics import accuracy_score
import warnings

warnings.filterwarnings('ignore')

filepath = '/content/drive/MyDrive/Machine Learning Dataset/place.csv'

dataframe = pd.read_csv(filepath)

dataframe.sample(5)
```

```python
x = dataframe.iloc[:, 0:3]
y = dataframe.iloc[:, 3]
```

```python
#similer to train_test_split
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
logisticregression = LogisticRegression()
score = cross_val_score(logisticregression, x, y, cv=10)

print("Cross Validation Scores are {}".format(score))
print("Average Cross Validation (Test data accuracy): {}".format(score.mean()))
```

```python
logisticregression1 = LogisticRegression()
score=cross_val_score(logisticregression1,x,y,cv=StratifiedKFold(10))

print("Cross Validation Scores are {}".format(score))
print("Average Cross Validation (Test data accuracy): {}".format(score.mean()))
```
