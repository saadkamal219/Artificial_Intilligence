#**Linear regression model**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataframe = pd.read_csv('/content/drive/MyDrive/Machine Learning Dataset/PriceDataSet.csv')

dataframe.head()
```

```python
dataframe.isnull().sum()
```

```python
dataframe.columns
```

```python
x = dataframe[['weight']].values
y = dataframe['price']
```

```python
x
```

```python
y
```

```python
plt.scatter(x, y, marker = 'o', color = 'red')
plt.title('Price and Weight Observation')
plt.xlabel('Weight')
plt.ylabel('Price')

plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression

pin = LinearRegression()
```

```python
pin.fit(x_train, y_train)
```

```python
pin.predict(x_test)
```

```python
y_test
```

```python
plt.plot(dataframe['weight'], pin.predict(dataframe[['weight']]))
```

```python
plt.scatter(x, y, marker = 'o', color = 'red')
plt.title('Reality vs Model Calculation')
plt.xlabel('Weight')
plt.ylabel('Price')
plt.plot(dataframe['weight'], pin.predict(dataframe[['weight']]))
```

```python
pin.score(x_test, y_test) * 100
```
