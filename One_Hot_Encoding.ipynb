```python
import pandas as pd
import numpy as np

filepath = '/content/drive/MyDrive/Machine Learning Dataset/customer.csv'

dataframe = pd.read_csv(filepath)

dataframe.sample(5)
```

```python
dataframe['gender'].value_counts()
```

```python
dummies = pd.get_dummies(dataframe, columns=['gender'])
dummies
```

```python
from sklearn.preprocessing import OneHotEncoder
import warnings

warnings.filterwarnings('ignore')

onehotencoder = OneHotEncoder(drop='first', sparse=False, dtype=np.int32)
dataframe['gender'] = onehotencoder.fit_transform(dataframe[['gender']])
```
