```python
import numpy as np
import pandas as pd

filepath = '/content/drive/MyDrive/Machine Learning Dataset/customer.csv'

dataframe = pd.read_csv(filepath)

dataframe.sample(5)
```

```python
dataframe.isnull().sum()
```

#Ordinal Encoding

```python
dataframe[['review', 'education']]
```

```python
dataframe['review'].unique()
```

```python
dataframe['education'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinalencoder = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
dataframe[['review', 'education']] = ordinalencoder.fit_transform(dataframe[['review', 'education']])
```
